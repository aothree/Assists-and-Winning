## Combine Historical Team Wins and Team Stats into one main dataframe `mdf`

```python
def clean_team_name(my_string):
    if '*' in my_string:
        return my_string[:-5].strip()
    else:
        return my_string[:-4].strip()
```

```python
def clean_team_name2(my_string):
    if '*' in my_string:
        return my_string[:-1].strip()
    else:
        return my_string.strip()
```

```python
years = ['2022','2021', '2020', '2019', '2018','2017', '2016', '2015', '2014', '2013', '2012', '2011', 
         '2010', '2009', '2008', '2007', '2006', '2005', '2004', '2003', '2002', '2001', '2000', 
         '1999', '1998', '1997', '1996', '1995', '1994', '1993', '1992', '1991', '1990',
        '1989', '1988', '1987', '1986', '1985', '1984', '1983', '1982', '1981', '1980']

for year in years:
    if year == '2022': 
        url = f'https://www.basketball-reference.com/leagues/NBA_{year}.html'
        res = requests.get(url)
        with open(f'./Data/html_files/NBA_{year}.xls', 'wb') as f:
            f.write(res.content)

        df = pd.read_html(f'./Data/html_files/NBA_{year}.xls')

        # get records 
        wins_df1 = pd.DataFrame(df[0]) 
        wins_df1.rename({'Eastern Conference': 'Team'}, axis = 1, inplace = True)
        wins_df1['Team'] = wins_df1['Team'].map(lambda x: clean_team_name(x))
        wins_df2 = pd.DataFrame(df[1]) 
        wins_df2.rename({'Western Conference': 'Team'}, axis = 1, inplace = True)
        wins_df2['Team'] = wins_df2['Team'].map(lambda x: clean_team_name(x))
        dfs = [wins_df1, wins_df2]
        wins_df = pd.concat(dfs, ignore_index = True)

        # get stats
        stats = pd.DataFrame(df[4]) 
        stats['Team'] = stats['Team'].map(lambda x: clean_team_name2(x))

        # merge and save to .csv
        df_merged = pd.merge(left = wins_df, right = stats, on = 'Team')
        df_merged['team_year'] = df_merged['Team'] + '_' + year
        df_merged['year'] = year
        df_merged.to_csv(f'./Data/yearly_stats/{year}_stats.csv')
        time.sleep(3)
    elif int(year) >2015 :
        url = f'https://www.basketball-reference.com/leagues/NBA_{year}.html'
        res = requests.get(url)
        with open(f'./Data/html_files/NBA_{year}.xls', 'wb') as f:
            f.write(res.content)

        df = pd.read_html(f'./Data/html_files/NBA_{year}.xls')

        # get records 
        wins_df1 = pd.DataFrame(df[0]) 
        wins_df1.rename({'Eastern Conference': 'Team'}, axis = 1, inplace = True)
        wins_df1['Team'] = wins_df1['Team'].map(lambda x: clean_team_name2(x))
        wins_df2 = pd.DataFrame(df[1]) 
        wins_df2.rename({'Western Conference': 'Team'}, axis = 1, inplace = True)
        wins_df2['Team'] = wins_df2['Team'].map(lambda x: clean_team_name2(x))
        dfs = [wins_df1, wins_df2]
        wins_df = pd.concat(dfs, ignore_index = True)

        # get stats
        stats = pd.DataFrame(df[4]) 
        stats['Team'] = stats['Team'].map(lambda x: clean_team_name2(x))

        # merge and save to .csv
        df_merged = pd.merge(left = wins_df, right = stats, on = 'Team')
        df_merged['team_year'] = df_merged['Team'] + '_' + year
        df_merged['year'] = year
        df_merged.to_csv(f'./Data/yearly_stats/{year}_stats.csv')
        time.sleep(3)
       
    else:
        print(f'scraping for {year}')
        url = f'https://www.basketball-reference.com/leagues/NBA_{year}.html'
        res = requests.get(url)
        with open(f'./Data/html_files/NBA_{year}.xls', 'wb') as f:
            f.write(res.content)

        df = pd.read_html(f'./Data/html_files/NBA_{year}.xls')

        # get records 
        wins_df1 = pd.DataFrame(df[0]) 
        wins_df1.rename({'Eastern Conference': 'Team'}, axis = 1, inplace = True)
        wins_df1['Team'] = wins_df1['Team'].map(lambda x: clean_team_name2(x))
        wins_df2 = pd.DataFrame(df[1]) 
        wins_df2.rename({'Western Conference': 'Team'}, axis = 1, inplace = True)
        wins_df2['Team'] = wins_df2['Team'].map(lambda x: clean_team_name2(x))
        dfs = [wins_df1, wins_df2]
        wins_df = pd.concat(dfs, ignore_index = True)

        # get stats
        stats = pd.DataFrame(df[3]) 
        stats['Team'] = stats['Team'].map(lambda x: clean_team_name2(x))

        # merge and save to .csv
        df_merged = pd.merge(left = wins_df, right = stats, on = 'Team')
        df_merged['team_year'] = df_merged['Team'] + '_' + year
        df_merged['year'] = year
        df_merged.to_csv(f'./Data/yearly_stats/{year}_stats.csv')
        time.sleep(3)
```

## Scrape passing statistics from nba.com

```python
url = 'https://stats.nba.com/stats/leaguedashptstats?College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&GameScope=&GameSegment=&Height=&LastNGames=0&LeagueID=00&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PerMode=PerGame&Period=0&PlayerExperience=&PlayerOrTeam=Team&PlayerPosition=&PtMeasureType=Passing&Season=2021-22&SeasonSegment=&SeasonType=Regular+Season&StarterBench=&TeamID=0&VsConference=&VsDivision=&Weight='
```

```python
headers  = {
    'Connection': 'keep-alive',
    'Accept': 'application/json, text/plain, */*',
    'x-nba-stats-token': 'true',
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36',
    'x-nba-stats-origin': 'stats',
    'Sec-Fetch-Site': 'same-origin',
    'Sec-Fetch-Mode': 'cors',
    'Referer': 'https://stats.nba.com/',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'en-US,en;q=0.9',
}
```

```python
seasons = ['2021-22', '2020-21', '2019-20', '2018-19', '2017-18', '2016-17', '2015-16', '2014-15', '2013-14']
for season in seasons:
    url = f'https://stats.nba.com/stats/leaguedashptstats?College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&GameScope=&GameSegment=&Height=&LastNGames=0&LeagueID=00&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PerMode=PerGame&Period=0&PlayerExperience=&PlayerOrTeam=Team&PlayerPosition=&PtMeasureType=Passing&Season={season}&SeasonSegment=&SeasonType=Regular+Season&StarterBench=&TeamID=0&VsConference=&VsDivision=&Weight='
    res = requests.get(url=url, headers=headers).json()
    stats = res['resultSets'][0]['rowSet']
    df = pd.DataFrame(stats)
    df.rename(columns = {1: 'team_abbr',
                         2: 'team',
                         3: 'G',
                         4: 'W',
                         5: 'L',
                         7: 'passes_made',
                         9: 'asst',
                         11: 'secondary_asst',
                         12: 'potential_asst',
                         }, inplace = True)
    df.drop(columns = [0,6,8,10,13,14,15,16], inplace = True)
    df['year'] = season
    df.to_csv(f'./Data/nba_com_passing_stats/passing_{season}.csv')
```

```python
path = 'Data/nba_com_passing_stats' # use your path
files = glob.glob(os.path.join(path, "*.csv"))

lst = []
for file in files:
    df = pd.read_csv(file, index_col=0)
    lst.append(df)

passing = pd.concat(lst, axis=0,)
passing
```
